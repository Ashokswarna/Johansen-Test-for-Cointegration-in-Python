```python
def load_file():
    
    df = pd.read_excel('external_regressors_Auto.xlsx')
    df.head()
    df.reset_index(drop=True)
    df = df.set_index('Date')
    df.columns = df.columns.str.replace(' ', '_')
    cols = df.columns
    
    return df, cols
```
